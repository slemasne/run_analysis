```python
import requests
from bs4 import BeautifulSoup
import os
import pandas as pd
import html5lib
```

```python
url = r'https://www.jpmorganchasecc.com/results/search.php?city_id=10&search=1&gender=m&year=2017'
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.text, "lxml")
table = soup.find('table', id="results")
```

```python
for row in table.find_all('tr')[1:]:
    col = row.find_all('td')
    print col
```

```python
place = []
bib = []
name = []
time = []
gender = []
company = []
```

```python
for row in table.find_all('tr')[1:]:
    col = row.find_all('td')
    
    column_0 = col[0].string.strip().encode()
    place.append(column_0)
    
    column_2 = col[2].string.strip().encode()
    bib.append(column_2)
    
    column_3 = col[3].string.strip()
    name.append(column_3)
    
    column_4 = col[4].string.strip()
    time.append(column_4)
    
    column_5 = col[5].string.strip()
    gender.append(column_5)
    
    column_6 = col[6].string
    company.append(column_6)
```

```python
columns = {'place':place,
          'bib':bib,
          'name':name,
          'time':time,
           'gender':gender,
          'company':company}
```

```python
df = pd.DataFrame(columns, index = place)
```
